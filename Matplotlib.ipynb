# MATPLOTLIB TUTORIAL

```python
from matplotlib import pylab

print(pylab.__version__)
```

```python
import numpy as np
import pandas as pd

x = np.linspace(0,10,25) # Create array of 25 elements range 0 and 10. numbers devided equally in 25 parts

y = x*x+2

# print(x)
# print(y)

print(np.array([x,y]).reshape(25,2))

pylab.plot(x,y,'g')
```

# Drawing a subgraph

```python
pylab.subplot(2,1,1) # The contents in bracket are rows,columns and indexes
pylab.plot(x,y,'g-*')
pylab.subplot(2,1,2)
pylab.plot(y,x,'r--')
```

```python
from matplotlib import pylab
import numpy as np
import pandas as pd
pylab.subplot(1,2,1) # The contents in bracket are rows,columns and quadrant
pylab.plot(x,y,'g-*')
pylab.subplot(1,2,2)
pylab.plot(y,x,'r--')
```

```python
from matplotlib import pylab as plt

fig = plt.figure()  # Create canvas

axes = fig.add_axes([0.5,0.1,0.8,0.8]) # Control the left, right, height, width of the canvas from 0 to 1

axes.plot(x,y,'r')
```

```python
# again we draw subgraphs

fig, axes = plt.subplots(nrows=1,ncols=3) # Submap is of 1 row and 3 columns
for ax in axes:
    ax.plot(x,y,'r')
```

```python
# We can draw a picture or graph inside another graph

fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) # Big axes
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) # Small axes

axes1.plot(x,y,'r')
axes2.plot(y,x,'g--')
```

```python
fig = plt.figure(figsize=(16,9), dpi = 300)  # New Graphic objects with canvas size and dpi stands for image quality

axes = fig.add_axes([0.5,0.1,0.8,0.8]) # Control the left, right, height, width of the canvas from 0 to 1

axes.plot(x,y,'r')
```

# Limits on Graphs
# Set the canvas grid and axis range

```python
fig, Axes = plt.subplots(1,3,figsize=(10,5))

Axes[0].plot(x,x**2,x,x**3,lw=2)
Axes[0].grid(True)              #Enable grid lines mentioned by square blocks
Axes[0].legend(["y=x**2","y=x**3"],loc=1)

Axes[1].plot(x,x*2,x,x*3,lw=2)
Axes[1].grid(True)
Axes[1].legend(["y=2x","y=3x"],loc=1)

Axes[2].plot(x,x**2,x,x**3,lw=2)
Axes[2].set_ylim([0,60])
Axes[2].set_xlim([2,5])
Axes[2].grid(True)
Axes[2].legend(["y=x**2","y=x**3"],loc=1)
```

# Other 2D Graphics

```python
n = np.array([0,1,2,3,4,5])

fig, axes = plt.subplots(1,4,figsize=(16,5))

axes[0].set_title('Scatter plot')
axes[0].scatter(x,x+0.5*np.random.randn(len(x)))

axes[1].set_title('Step plot')
axes[1].step(n,n**2,lw=2)

axes[2].set_title('Bar plot')
axes[2].bar(n,n**2,align='center',alpha=0.5)

axes[3].set_title('fill_between')
axes[3].fill_between(x,x**2,x**3,color='g',alpha=0.8)  # It will fill the color between graphs of two functions
```

```python
## Draw a radar chart
import numpy as np
fig = plt.figure(figsize=(6,6))
ax = fig.add_axes([0,0,0.8,0.8],polar=True)

t = np.linspace(0, 2*np.pi, 100)
ax.plot(t,t,color='r',lw=3)
ax.plot(t,t*5,color='b',lw=3)
```

```python
# Draw Histogram

n = np.random.randn(10000)
fig, axes = plt.subplots(1,2,figsize=(12,4))

axes[0].set_title("Default Histogram")
axes[0].hist(n)

axes[1].set_title("Cumulative Detailed Histogram")
axes[1].hist(n, bins=15, cumulative=True)
```

# Draw Contour image

```python
import matplotlib
import numpy as np
import pandas as pd
import matplotlib.cm as cm    ## cm stands for color maps
import matplotlib.pyplot as plt

delta = 0.025
x = np.arange(-3.0,3.0,delta)
y = np.arange(-2.0,2.0,delta)

X,Y = np.meshgrid(x,y)

Z1 = np.exp(-X**2-Y**2)
Z2 = np.exp(-(X-1)**2-(Y-1)**2)
Z = (Z1 - Z2)*2
print(X)
print(Y)
print(Z)
```

```python
fig, ax = plt.subplots()
CS = ax.contour(X,Y,Z)

ax.clabel(CS,inline=1, fontsize=10)
ax.set_title('Contour Map')
```
